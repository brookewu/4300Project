```python
import pandas as pd
import numpy as np

# from scipy.cluster.vq import kmeans

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from tqdm import tqdm

import seaborn as sns
```

```python
jaccard = pd.read_csv("jaccard.csv")
jaccard = jaccard.drop("Unnamed: 0", axis=1)
jaccard.index = jaccard.columns
jaccard
```

```python
cosine = pd.read_csv("cosine_matrix_2.csv")
cosine = cosine.drop("Unnamed: 0", axis=1)
cosine.index = jaccard.index
cosine.columns = jaccard.columns
cosine
```

```python
svd = pd.read_csv("svd_matrix.csv")
svd = svd.drop("Unnamed: 0", axis=1)
svd.index = jaccard.index
svd.columns = jaccard.columns
svd
```

```python
first = svd
first_label = "first"

second = jaccard
second_label = "second"

third = cosine
third_label = "third"

restaurants = jaccard.index
filter_size = 100
```

```python
first_filter = []
for i in tqdm(range(len(first))):
    row = first.iloc[i]
    row = row.sort_values(ascending=False)
    row = row.drop(restaurants[i])
    for company in row.index[:filter_size]:
        first_filter.append((restaurants[i], company, first[company].iloc[i]))
filtered = pd.DataFrame(first_filter, columns=["x", "y", first_label])
filtered
```

```python
second_score = []
third_score = []
for i in tqdm(range(len(filtered))):
    row = filtered.iloc[i]
    second_score.append(second[row["x"]][row["y"]])
    third_score.append(third[row["x"]][row["y"]])
filtered[second_label] = second_score
filtered[third_label] = third_score
filtered
```

```python
filtered.to_csv("output.csv")
```
