```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
data = pd.read_csv("sample.csv")
data = data.drop(["Unnamed: 0","funny_user","cool_user","latitude","city", "state", "is_open", "date","attributes",\
                  "longitude","funny", "cool","friends", "compliment_hot", "compliment_more", "compliment_profile"\
                  , "compliment_cute", "compliment_list", "compliment_note", "compliment_plain", "compliment_cool"\
                 , "compliment_writer", "compliment_photos", "yelping_since"], axis = 1)

data
```

```python
business_stars = {}

for index, row in data.iterrows():
    # Get the business name and categories for this row
    business_name = row['name_user']
    categories = row['categories']
    star = row['stars_review']
    # Add the mapping to the dictionary
    if business_name in business_stars:
         business_stars[business_name] = (business_stars[business_name] + star)/2
    else:
        business_stars[business_name] = star

count5 = 0
count4 = 0
count3 = 0
count2 = 0
count1 = 0

for item in business_stars:
    star = business_stars[item]
    if (star) == 5.0:
        count5 += 1
    elif (star) == 4.0:
        count4 += 1
    elif (star) == 3.0:
        count3 += 1
    elif (star) == 2.0:
        count2 += 1
    elif (star) == 1.0:
        count1 += 1
print("5 star restaurants:" + str(count5))
print("4 star restaurants:" + str(count4))
print("3 star restaurants:" + str(count3))
print("2 star restaurants:" + str(count2))
print("1 star restaurants:" + str(count1))

reviewers = {}

for index, row in data.iterrows():
    reviewer = row['name']
    star = row['stars_review']
    if reviewer in reviewers:
        reviewers[reviewer] = (reviewers[reviewer][0] + 1, (reviewers[reviewer][1] + star) / 2)
    else:
        reviewers[reviewer] = (1,star)
        
reviewers = dict(sorted(reviewers.items(), key=lambda item: item[1], reverse = True))
print("Top 10 reviewers and their average ratings")
print("=================================================")
for item in reviewers:
    values = reviewers[item]
    print(item + ", reviews: " + str(values[0]) + ", average rating: " + str(values[1]))
```
