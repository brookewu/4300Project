```python
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
data = pd.read_csv("data.csv")
data
```

```python
dataset = data[["name", "categories"]].drop_duplicates("name")
dataset
```

```python
cats = {}
for i in range(len(dataset)):
    row = dataset.iloc[i]
    if pd.isna(row["categories"]):
        continue
    categories = set()
    for category in row["categories"].split(","):
        categories.add(category.strip())
    cats[row["name"]] = categories
```

```python
for company in list(cats.keys()):
    if "Restaurants" not in cats[company]:
        cats.pop(company)
```

```python
jaccard = np.zeros((len(cats), len(cats)))
companies = sorted(list(cats.keys()))

for x in tqdm(range(len(companies))):
    for y in range(len(companies)):
        jaccard[x, y] = len(cats[companies[x]].intersection(cats[companies[y]])) / len(cats[companies[x]].union(cats[companies[y]]))
```

```python
jac_df = pd.DataFrame(jaccard)
jac_df
```
